```python
def  isofreq_plot(file):
    
    f1 = file
    
    import pandas as pd
    import matplotlib

    fig, ax = plt.subplots(figsize=(14,10))
    title = 'Isofrequency diagram in momentum space'

    light_speed = 3e8

    df = pd.DataFrame(dict(x=f1[:-1,1], y=f1[:-1,2], z=f1[:-1,-1]*light_speed/(a*10**3))) #since a is in nm and we want THz
    xcol, ycol, zcol = "x", "y", "z"
    df = df.sort_values(by=[xcol, ycol])
    xvals = df[xcol].unique()
    yvals = df[ycol].unique()
    zvals = df[zcol].values.reshape(len(xvals), len(yvals)).T
    CS = ax.contour(xvals, yvals, zvals, levels = 45)
    ax.set_title(title)
    ax.set_xlabel('Normalized $k_x \, a / 2\pi$ ')
    ax.set_ylabel('Normalized $k_y \, a / 2\pi$ ')
    norm= matplotlib.colors.Normalize(vmin=CS.cvalues.min(), vmax=CS.cvalues.max())
    sm = plt.cm.ScalarMappable(norm=norm, cmap = CS.cmap)
    sm.set_array([])
    fig.colorbar(sm, label='$\\nu$ (THz)')

    plt.text(0.015, 0.02, '$\\Gamma$',bbox=dict(facecolor='gray', alpha=0.9, boxstyle='Round'), fontsize=24, color='white')
    plt.text(0.015, 0.47, '$Y$',bbox=dict(facecolor='gray', alpha=0.9, boxstyle='Round'), fontsize=24, color='white')
    plt.text(0.47, 0.47, '$M$',bbox=dict(facecolor='gray', alpha=0.9, boxstyle='Round'), fontsize=24, color='white')
    plt.text(0.47, 0.02, '$X$',bbox=dict(facecolor='gray', alpha=0.9, boxstyle='Round'), fontsize=24, color='white')
    
    return fig, ax
```

## Slope walls

Let's see what happens when we include the slope walls:

```python
def eps_func (Vector3):
    array3 = np.array(Vector3)
    x = array3 [0]
    y = array3 [1]
    z = array3 [2]
    

    dielec_const = 1.9935 # this can be modified
    
    h = t #  nanometers
    alpha = 7.9 * np.pi / 180 # degrees to radians
    r1 = r #  upper radius (biggest one)
    r2 = r1 - h * np.sin(alpha)
    
        
    x_prime = 0.5*(r1+r2)+ z*(r1-r2)/h #this defines the slope of the walls
    y_prime = x_prime                  #the slope is the same in both directions (simmetry)
    
    if x**2 + y**2 > x_prime**2 and np.abs(z) < h*0.5:
        
        eps = mp.Medium(index = dielec_const) 
        
    else:
        
        eps = mp.Medium(index = 1) 
        
    return eps
```

```python
resolution = 24
default_material = eps_func
```

```python
if plot == True:
    sim = mp.Simulation(resolution=resolution,
                    cell_size=mp.Vector3(sx,sy,sz),
                    material_function=default_material)

    sim.init_sim()

    eps_data = sim.get_epsilon()
```

```python
if plot == True:
    print(np.shape(eps_data))
```

```python
if plot == True:
    fig, ax = plt.subplots(1,2,figsize=(14,6))
    ax1, ax2 = ax
    fig.suptitle('Cross sections of the 2D Photonic Crystal  Slab')
    pos1 = ax1.imshow(eps_data[:,:,120], interpolation='spline36', cmap='binary')
    ax1.axis('off')
    pos2 = ax2.imshow(np.rot90(eps_data[:,12,100:140],2).T, interpolation='spline36', cmap='binary')
    ax2.axis('off')
    fig.colorbar(pos2, ax=ax1)
```

```python
import pandas as pd
import matplotlib
from mpl_toolkits.axes_grid1 import make_axes_locatable


fig, ax = plt.subplots(2,1,figsize=(10,18))
ax1, ax2 = ax
title = 'Isofrequency diagram in momentum space'

light_speed = 3e8

for file, axis in zip([f1,f5],[ax1,ax2]):
    df = pd.DataFrame(dict(x=file[:-1,1], y=file[:-1,2], z=file[:-1,-1]*light_speed/(a*10**3)))
    xcol, ycol, zcol = "x", "y", "z"
    df = df.sort_values(by=[xcol, ycol])
    xvals = df[xcol].unique()
    yvals = df[ycol].unique()
    zvals = df[zcol].values.reshape(len(xvals), len(yvals)).T
    CS = axis.contour(xvals, yvals, zvals, levels = 45)
    axis.set_xlabel('Normalized $k_x \, a / 2\pi$ ')
    axis.set_ylabel('Normalized $k_y \, a / 2\pi$ ')
    
    divider = make_axes_locatable(axis)
    cax = divider.append_axes('right', size='5%', pad=0.5)

    
    norm= matplotlib.colors.Normalize(vmin=CS.cvalues.min(), vmax=CS.cvalues.max())
    sm = plt.cm.ScalarMappable(norm=norm, cmap = CS.cmap)
    sm.set_array([])
    fig.colorbar(sm, cax=cax, orientation='vertical', label='$\\nu$ (THz)')

    axis.text(0.02, 0.03, '$\\Gamma$',bbox=dict(facecolor='gray', alpha=0.9, boxstyle='Round'), fontsize=24, color='white')
    axis.text(0.02, 0.46, '$Y$',bbox=dict(facecolor='gray', alpha=0.9, boxstyle='Round'), fontsize=24, color='white')
    axis.text(0.46, 0.46, '$M$',bbox=dict(facecolor='gray', alpha=0.9, boxstyle='Round'), fontsize=24, color='white')
    axis.text(0.46, 0.03, '$X$',bbox=dict(facecolor='gray', alpha=0.9, boxstyle='Round'), fontsize=24, color='white')
    
ax[0].set_title(title+str(' with no slope'))
ax[1].set_title(title+str(' with a 7.9 slope'))
```
